```python
#%matplotlib ipympl
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import re
import math
import statsmodels.api as sm
import warnings
warnings.filterwarnings('ignore')

def ols_fit(depend, independ):

    independ_with_constant = sm.add_constant(independ)
    model = sm.OLS(depend, independ_with_constant).fit()

    return model.summary()


def sum_number_by_split_char(s, split_char):

    bonus = s.replace('ToF', '20MB')
    bonus = bonus.replace('3D', '30MB')

    list_number = [re.findall('\d+', x)[0] for x in bonus.split(split_char)]

    parsed_number = [float(x) for x in list_number]
    return sum(parsed_number)




# Load the dataset
file_path = 'mobile.csv'
new_mobile_data = pd.read_csv(file_path)
new_mobile_data.columns = new_mobile_data.columns.str.strip()
```

```python

def describe_price_data(d):
    data_ensured_number = d.replace('[\$,]','', regex=True).astype(float)
    n, minmax, mean, variance, skewness, _ = stats.describe(data_ensured_number)

    #Estimation
    standard_deviation = math.sqrt(variance)
    print(f'Describe Price:\nn={n}\nmin={minmax[0]}, max={minmax[1]}\nmean={mean}\nvariance={variance}\nstandard deviation={standard_deviation}\nskewness={skewness}')

    his = plt.figure(1) 
    plt.hist(data_ensured_number, bins=30)
    plt.title('Distribution of price')
    plt.xlabel('Price ($)')
    plt.ylabel('Frequency' )
    plt.grid(True)

    his.show()

    box = plt.figure(2)
    plt.boxplot(data_ensured_number, vert=False, patch_artist=True)
    plt.title('Box plot of Price')
    plt.xlabel('Price ($)')
    plt.grid(True)

    box.show()

describe_price_data(new_mobile_data['Price ($)'])
```

- Interquartile Range (IQR): The prices at the 25th percentile (\\$99) and the 75th percentile (\\$499) show that the middle 50% of the phones are priced within this range. The IQR of \$300 indicates a moderate spread among the central 50% of the data.

Outliers: Given the maximum price of \$1999, there are likely a few high-end smartphones that are significantly more expensive than the rest. These outliers affect the mean, making it higher than the median.

# Affordability:

# Market Segmentation:

```python
#set dependent 
#remove $ , 
y = new_mobile_data['Price ($)'].replace('[\$,]','', regex=True).astype(float) 
```

## Set alpha = 0.05

# Hypothesis 1: Screen Size and Price
- Null Hypothesis (H0): There is no significant relationship between screen size and the price of mobile phones.
- Alternative Hypothesis (H1): There is a significant relationship between screen size and the price of mobile phones.


```python
#6.8 + 1.3 => 6.8 
x_screen = new_mobile_data['Screen Size (inches)'].str.extract(r'^(\d+.?\d*)').astype(float)
model_screen_summary = ols_fit(y, x_screen)
print(model_screen_summary)
```

Summary:
- P-value: 0.351 (not significant)
- Conclusion: There is no significant relationship between screen size and price.

# Hypothesis 2: RAM and Price
- Null Hypothesis (H0): There is no significant relationship between the amount of RAM and the price of mobile phones.
- Alternative Hypothesis (H1): There is a significant relationship between the amount of RAM and the price of mobile phones.

```python
x_ram = new_mobile_data['RAM'].str.extract(r'(\d+)').astype(float)
model_ram_summary = ols_fit(y, x_ram)
print(model_ram_summary)
```

Summary:
- P-value: < 0.05 (significant)
- Conclusion: There is a significant relationship between the amount of RAM and the price of mobile phones.

# Hypothesis 3: Storage Capacity and Price
- Null Hypothesis (H0): There is no significant relationship between storage capacity and the price of mobile phones.
- Alternative Hypothesis (H1): There is a significant relationship between storage capacity and the price of mobile phones.

```python
x_storage = new_mobile_data['Storage'].str.extract(r'(\d+)').astype(float)
model_storage_summary = ols_fit(y, x_storage)
print(model_storage_summary)
```

Summary:
- P-value: < 0.05 (significant)
- Conclusion: There is a significant relationship between storage capacity and the price of mobile phones.

# Hypothesis 4: Camera Quality and Price
- Null Hypothesis (H0): There is no significant relationship between the camera quality (measured in megapixels) and the price of mobile phones.
- Alternative Hypothesis (H1): There is a significant relationship between the camera quality (measured in megapixels) and the price of mobile phones.

```python
x_camera = [sum_number_by_split_char(s, '+') for s in new_mobile_data['Camera (MP)']]
model_camera_summary = ols_fit(y, x_camera)
print(model_camera_summary)
```

Summary:
- P-value: 0.085 (not significant)
- Conclusion: There is no significant relationship between camera quality and price.

# Price and Ram linear regression

```python
def linear_regression(x, y, prediction, xlabel, ylabel):
    
    plt.figure(figsize=(10, 6))
    plt.scatter(x=x, y=y, label=ylabel)
    plt.plot(x, prediction, color='red', label='Regression Line')
    plt.title(f'Regression of {ylabel} on {xlabel}')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True)
    plt.show()

x_ram_with_constant = sm.add_constant(x_ram)
model = sm.OLS(y, x_ram_with_constant).fit()
price_prediction = model.predict(x_ram_with_constant)

print(model.summary())
linear_regression(x_ram, y, price_prediction, 'Ram', 'Price')
print(f'linear regresion equation   y = {model.params['const']} + {model.params[0]}*x')
```

```python
import numpy as np

def linear_regression_3d(x1, x2, y, predicted, x1label, x2label, ylabel):    

    fig = plt.figure()

    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(x1, x2, y, marker = '.', color='blue')
    ax.set_xlabel(x1label)
    ax.set_ylabel(x2label)
    ax.set_zlabel(ylabel)
    
    ax.plot_trisurf(x1, x2, predicted, alpha=0.5)
    plt.show()

    

x = [x_ram[0], x_storage[0]]
def reg_m(y, x):
    ones = np.ones(len(x[0]))
    X = sm.add_constant(np.column_stack((x[0], ones)))
    for ele in x[1:]:
        X = sm.add_constant(np.column_stack((ele, X)))
    results = sm.OLS(y, X).fit()
    return results, X



price_by_ram_storage_model, x_train = reg_m(y, x)
print(price_by_ram_storage_model.summary())
price_predicted_by_3d_model = price_by_ram_storage_model.predict(x_train)
print(f'linear regression Price with 2 variables (Ram, Storage):\ny = {price_by_ram_storage_model.params['const']} + {price_by_ram_storage_model.params['x1']}*x1 + {price_by_ram_storage_model.params['x2']}*x2')  
linear_regression_3d(x_train[:,0], x_train[:,1], y, price_predicted_by_3d_model, 'Ram', 'Storage', 'Price')
#linear_regression(x_train, y, predict_price, 'Ram + Storage', 'Price')
```
